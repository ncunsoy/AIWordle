```python

import random
import pandas as pd
import numpy as np


words=pd.read_csv('words.csv')
words['First'] = list(w[0] for w in words['word'])
words['Second'] = list(w[1] for w in words['word'])
words['Third'] = list(w[2] for w in words['word'])
words['Fourth'] = list(w[3] for w in words['word'])
words['Fifth'] = list(w[4] for w in words['word'])

class AI_Player:
    alpha = [chr(i) for i in range(ord('a'), ord('z') + 1)]
    columnList = ['First', 'Second', 'Third', 'Fourth', 'Fifth']
    wordList=[]
    gWord = "aaaaa"
    playerRound = 1

    def __init__(self):
        self.wordList = pd.read_csv('words.csv')
        self.wordList['First'] = list(w[0] for w in self.wordList['word'])
        self.wordList['Second'] = list(w[1] for w in self.wordList['word'])
        self.wordList['Third'] = list(w[2] for w in self.wordList['word'])
        self.wordList['Fourth'] = list(w[3] for w in self.wordList['word'])
        self.wordList['Fifth'] = list(w[4] for w in self.wordList['word'])
        

    def calculateProbability(self):
        if (self.wordList.shape[0] < 900):
            letter1 = {}
            letter2 = {}
            letter3 = {}
            letter4 = {}
            letter5 = {}
            dictList = [letter1, letter2, letter3, letter4, letter5]

            for i in range(0, 5):
                for letter in self.alpha:
                    num = self.wordList.loc[self.wordList[self.columnList[i]] == letter].shape[0] - 1
                    dictList[i].update({letter: num})

            for i in range(0, 5):
                self.wordList[f'Frequency{i + 1}'] = [dictList[i].get(k) for k in self.wordList[self.columnList[i]]]
                self.wordList[f'Frequency{i + 1}'] = self.wordList[f'Frequency{i + 1}'] / (self.wordList.shape[0] - 1)

            self.wordList['Complex Probability Value'] = (self.wordList['Frequency1'] + self.wordList['Frequency2'] +
                                                        self.wordList['Frequency3'] +
                                                        self.wordList['Frequency4'] + self.wordList['Frequency5']) / 5
            self.wordList = self.wordList.sort_values(by='Complex Probability Value', ascending=False)


        else:
            allLetters = {}
            for letter in self.alpha:
                num = 0
                for i in range(0, 5):
                    num += self.wordList.loc[self.wordList[self.columnList[i]] == letter].shape[0] - 1
                allLetters.update({letter: num})

            def calculateBasicProbability(stringList):
                newList = []
                for string in stringList:
                    charSet = set()
                    score = 0
                    for char in string:
                        charSet.add(char)

                    for ch in charSet:
                        score += allLetters.get(f'{ch}')
                    newList.append(score)
                return newList

            self.wordList['Basic Probability Value'] = calculateBasicProbability(self.wordList['word'])
            self.wordList = self.wordList.sort_values(by='Basic Probability Value', ascending=False)
        return self.wordList

    def play(self, feedback: list):
        if (self.playerRound > 1):
            self.filterWords(feedback)
            self.calculateProbability()
            self.gWord = self.guess()

        else:
            self.calculateProbability()
            self.gWord = self.guess()
        self.wordList.drop(self.wordList.index[(self.wordList["word"] == self.gWord)],axis=0,inplace=True)
        self.playerRound+=1
        return self.gWord

    def guess(self):
        justWords=list (self.wordList['word'])
        self.gWord=str(justWords[0])
        return str(justWords[0])

    def filterWords(self, feedback: list):
        for i in range(0, len(feedback)):
            if (feedback[i] == "G"):  # GREEN
                self.wordList = self.wordList.loc[self.wordList[self.columnList[i]] == self.gWord[i]]
            elif (feedback[i] == "Y"):  # YELLOW
                self.wordList = self.wordList.loc[
                    (self.wordList['First'] == self.gWord[i]) | (self.wordList['Second'] == self.gWord[i])|
                    (self.wordList['Third'] ==self.gWord[i]) | (self.wordList['Fourth'] == self.gWord[i]) | (self.wordList['Fifth'] == self.gWord[i])
                    ]
            else:
                self.wordList = self.wordList.loc[
                    (self.wordList['First'] != self.gWord[i]) & (self.wordList['Second'] != self.gWord[i]) &
                    (self.wordList['Third'] !=self.gWord[i]) & (self.wordList['Fourth'] != self.gWord[i]) & (self.wordList['Fifth'] != self.gWord[i])
                    ]
        return self.wordList


class Game:
    wordListGame=[]
    word="" 
    gameRound = 6
    player=None
    feedback = []
    score=0

    def __init__(self):
        self.player=AI_Player()
        self.wordListGame=pd.read_csv('words.csv')
        self.word = np.random.choice(words['word'], 1)[0]
        self.score=0

    def startGame(self):
        print("\n", self.word, "\n")
        while (self.gameRound > 0):
            guess = self.player.play(self.feedback)
            self.feedback=[]
            print(guess)
            if self.word == guess:
                self.feedback == ['G', 'G', 'G', 'G', 'G']
                print("AI wins")
                print(self.feedback)
                return True

            else:
                for i in range(0, len(guess)):
                    if guess[i] == self.word[i]:
                        self.feedback.append('G')
                    elif guess[i] in self.word:
                        self.feedback.append('Y')
                    else:
                        self.feedback.append('W')
                print(self.feedback)

            self.gameRound -= 1
            if(self.gameRound==0):
                self.score=self.calculateScore()
        return False

    def returnAnswer(self):
        return self.word

    def calculateScore(self):
        score=0
        for letter in self.feedback:
            if letter=='G':
                score+=2
            elif letter=='Y':
                score+=1
        return score 
```
